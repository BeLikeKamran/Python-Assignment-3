```python
# 7. Write a program which takes 5 inputs from user for different subjectâ€™s marks, total it  and generate mark sheet using grades ?  

English=int(input("Enter marks of English:"))
Urdu=int(input("Enter marks of Urdu:"))
Maths=int(input("Enter marks of Maths:"))
Islamiyat=int(input("Enter marks of Islamiyat:"))
Computer=int(input("Enter marks of Computer:"))
per =(English+Urdu+Maths+Islamiyat+Computer)/3

if(per>=80):
    print("A+")
    
elif(per>=70):
    print("A")
    
elif(per>=60):
    print("B")
    
elif(per>=50):
    print("C")
    
elif(per>=40):
    print("D")
    
elif(per<40):
    print("Fail")
    
else:
    print("Invalid input")
```
